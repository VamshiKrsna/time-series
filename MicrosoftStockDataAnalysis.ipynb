# Stationarity Test

```python
tsdata.head()
```

# Testing the stationarity of Closing prices. 

```python
from statsmodels.tsa.stattools import adfuller
close_data = tsdata['Close']
```

```python
result = adfuller(close_data)
```

```python
print('ADF Statistic:', result[0])
print('p-value:', result[1])
print('Critical Values:', result[4])
if result[1] <= 0.05:
    print('likely stationary.')
else:
    print('likely non-stationary.')
```

Hence, Data is Non Stationary

# Rolling mean, std

```python
numeric_columns = tsdata.select_dtypes(include=['float64', 'int64'])

moving_mean = numeric_columns.rolling(3).mean()
moving_std = numeric_columns.rolling(3).std()

rolling_results = pd.concat([tsdata['Date'], moving_mean, moving_std], axis=1)

print(rolling_results)
```

```python
moving_std
```

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize = (8,4))
plt.plot(tsdata['Date'], tsdata['Close'])
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.title('Microsoft Stock Prices Over Time')
plt.xticks(rotation = 45)
plt.show()
```

```python
orig = plt.plot(tsdata['Date'], tsdata['Close'], color='blue', label='Original')

mean = plt.plot(rolling_results['Date'], moving_mean['Close'], color='red', label='Rolling Mean')
std = plt.plot(rolling_results['Date'], moving_std['Close'], color='black', label='Rolling Std')

plt.legend(loc='best')
plt.title('Rolling Mean & Standard Deviation')
plt.xlabel('Date')
plt.ylabel('Stock Price')
plt.xticks(rotation = 30)

plt.show(block=False)
```
